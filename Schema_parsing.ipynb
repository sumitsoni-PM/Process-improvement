```python
import json
import pandas as pd


schema = json.loads(json_string)

# GET PRIMARY KEY & OPEN TIMESTAMP (most imp)
ax_map = schema.get("axiontology_mapping", {})

primary_key = ax_map.get("_id")
open_timestamp = ax_map.get("openedTimestamp")

## here is all columsn information
transformations = schema.get("preprocessing_transformations", [])

rows = []

for t in transformations:
    col = t.get("attribute", "").strip()
    dtype = t.get("data_type", "UNKNOWN")
    fmt = t.get("formatter_kwargs", {})

    datatype_final = dtype
    additional_info = []
    special_details = []

    # LIST DATATYPE HANDLING
    if dtype == "LIST":
        parsing_method = fmt.get("parsing_method")
        delimiter = fmt.get("delimiter")

        if parsing_method == "ARRAY_SYNTAX":
            datatype_final = "LIST - Array"

        elif parsing_method == "SPLIT_ON_DELIMITER":
            if delimiter == ",":
                datatype_final = "LIST - Comma Separated"
            elif delimiter == "$":
                datatype_final = "LIST - Dollar Separated"
            else:
                datatype_final = "LIST - Delimiter Separated"
        else:
            datatype_final = "LIST"

    elif dtype == "STRING":
        datatype_final = "STRING"

    elif dtype == "NUMBER":
        datatype_final = "NUMBER"

    elif dtype == "DATE":
        datatype_final = "DATE"

    # any aditional information from formatter_kwargs
    for k, v in fmt.items():
        if isinstance(v, dict):
            for sub_k, sub_v in v.items():
                additional_info.append(f"{k}.{sub_k}={sub_v}")
        else:
            additional_info.append(f"{k}={v}")

    # -----------------------------
    # SPECIAL DETAILS
    # -----------------------------
    if primary_key and col == primary_key:
        special_details.append("Primary Key")

    if open_timestamp and col == open_timestamp:
        special_details.append("Open Timestamp")

    # -----------------------------
    # FINAL ROW
    # -----------------------------
    rows.append({
        "Column Name": col,
        "Datatype": datatype_final,
        "Special Details": ", ".join(special_details),
        "Additional Information": ", ".join(additional_info)
    })

df = pd.DataFrame(rows)
df.to_clipboard(index=False)
```
